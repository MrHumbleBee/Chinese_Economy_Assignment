```python
#人口规模与gdp
#数据清洗与可视化
#数据来源Wind,2018-12
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np 
plt.rcParams['font.sans-serif']=['SimHei']
import seaborn as sns

```

```python
city_pop=city_pop.drop([10,35,38,39,63,64])
```

```python
city_gdp=list(city_pop['GDP\n亿元'])
city_pop1=list(city_pop['population\n万人'])
txt=list(city_pop['city'])
```

```python
fig,ax=plt.subplots(figsize=(10,10))
grid=plt.GridSpec(4,4,wspace=0.7,hspace=0.7)
fig.text(x=0.3,y=0.05,s='人口总量与GDP的关系及边缘密度直方图',fontsize=16,fontweight='bold',
         verticalalignment='bottom')
bins=66

main_ax=plt.subplot(grid[1:4,1:4])

plt.plot(city_pop1,city_gdp,'ok',markersize=6,alpha=0.7)
plt.annotate('上海',xy=(2423.78,36011.82))
plt.annotate('北京',xy=(2154.20,33106))
plt.annotate('深圳',xy=(1302.66,24221.98))
plt.annotate('广州',xy=(1490.44,22859.35))
plt.xlabel('人口（万人）')
plt.ylabel('GDP（亿元）')
plt.grid(ls='--')
z1=np.polyfit(city_pop1,city_gdp,1)
p1=np.poly1d(z1)
print(p1)
yvals=p1(city_pop1)
plt.plot(city_pop1, yvals,color='gray',lw=4,alpha=0.5)
#gridobj = sns.lmplot(x='GDP\n亿元', y='population\n万人',data=city_pop, 
#                      robust=True)

y_hist=plt.subplot(grid[1:4,0],sharey=main_ax)
plt.hist(city_gdp,bins,orientation='horizontal',color='gray',alpha=1)
y_hist.invert_xaxis()
y_hist.plot(kind='kde',color='red')

x_hist=plt.subplot(grid[0,1:4],sharex=main_ax)
plt.hist(city_pop1,bins,orientation='vertical',color='gray',alpha=1)
#x_hist.invert_yaxis()
plt.savefig(r'C:\Users\PP\Desktop\数据\数据\中国主要城市人口与gdp关系的边缘密度直方图.png')
plt.show()
```

```python
fig,ax=plt.subplots(2,2,figsize=(20,16),dpi=100)

#子图1：平均寿命

China=list(life_Ex['中国'])
Japan=list(life_Ex['日本'])
America=list(life_Ex['美国'])
England=list(life_Ex['英国'])
date=list(life_Ex['国家'])

colors=['#e41a1c','#377eb8','#4daf4a','#984ea3']
markers=['o','s','H','D']    
labels=['中国','美国','日本','英国']

#fig,ax=plt.subplots(1,1,figsize=(10.,8),dpi=100)
    
ax[0,0].plot(date,China,
         marker=markers[0], markerfacecolor=colors[0], markersize=4,markeredgewidth=0.5,
         color="k", linewidth=0.8, linestyle="-",label=labels[0])
ax[0,0].plot(date,America,
         marker=markers[1], markerfacecolor=colors[1], markersize=4,markeredgewidth=0.5,
         color="k", linewidth=0.8, linestyle="-",label=labels[1])
ax[0,0].plot(date,Japan,
        marker=markers[2], markerfacecolor=colors[2], markersize=4,markeredgewidth=0.5,
         color="k", linewidth=0.8, linestyle="-",label=labels[2])
ax[0,0].plot(date,England,
        marker=markers[3], markerfacecolor=colors[3], markersize=4,markeredgewidth=0.5,
         color="k", linewidth=0.8, linestyle="-",label=labels[3])
ax[0,0].grid(linestyle='--')

ax[0,0].set_xlabel('国家',fontsize=24)
ax[0,0].set_ylabel('平均寿命',fontsize=24)

ax[0,0].legend(loc='lower right',edgecolor='none',facecolor='none',fontsize=18)
ax[0,0].spines['right'].set_color('none')
ax[0,0].spines['top'].set_color('none')
plt.xticks(fontsize=44)
plt.yticks(fontsize=33)

#ax[0,0].text(x=1,y=0,s='主要国家人口寿命对比',fontsize=16,fontweight='bold')


#子图2：公里里程对比
year1=list(Road_comp['年份'])
Chinese_road=list(Road_comp['中国'])
American_road=list(Road_comp['美国1'])

ax[0,1].plot(year1,Chinese_road,
             marker=markers[0], markerfacecolor=colors[0], markersize=4,markeredgewidth=0.5,
         color="k", linewidth=0.8, linestyle="-",label=labels[0])
ax[0,1].plot(year1,American_road,
             marker=markers[1], markerfacecolor=colors[1], markersize=4,markeredgewidth=0.5,
         color="k", linewidth=0.8, linestyle="-",label=labels[1])
ax[0,1].set_xlabel('年份',fontsize=24)
ax[0,1].set_ylabel('公路总里程（万公里）',fontsize=24)
ax[0,1].legend(loc='lower right',edgecolor='none',facecolor='none',fontsize=18)
ax[0,1].spines['right'].set_color('none')
ax[0,1].spines['top'].set_color('none')
plt.xticks(fontsize=33)
plt.yticks(fontsize=33)
ax[0,1].grid(linestyle='--')

#子图3：各国工业增加值对比
China1=list(Manufacturing_comp['中国'])
America1=list(Manufacturing_comp['美国'])
Japan1=list(Manufacturing_comp['日本'])
England1=list(Manufacturing_comp['英国'])
date1=list(Manufacturing_comp['国家'])

ax[1,0].plot(date1,China1,
           marker=markers[0], markerfacecolor=colors[0], markersize=4,markeredgewidth=0.5,
         color="k", linewidth=0.8, linestyle="-",label=labels[0] )
ax[1,0].plot(date1,Japan1,
           marker=markers[2], markerfacecolor=colors[2], markersize=4,markeredgewidth=0.5,
         color="k", linewidth=0.8, linestyle="-",label=labels[2] )
ax[1,0].plot(date1,America1,
             marker=markers[1], markerfacecolor=colors[1], markersize=4,markeredgewidth=0.5,
         color="k", linewidth=0.8, linestyle="-",label=labels[1])
ax[1,0].plot(date1,England1,
            marker=markers[3], markerfacecolor=colors[3], markersize=4,markeredgewidth=0.5,
         color="k", linewidth=0.8, linestyle="-",label=labels[3])
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
ax[1,0].legend(loc='lower left',edgecolor='none',facecolor='none',fontsize=18)
ax[1,0].set_xlabel('年份',fontsize=24)
ax[1,0].set_ylabel('工业增加值（百分比）',fontsize=24)
ax[1,0].spines['right'].set_color('none')
ax[1,0].spines['top'].set_color('none')
ax[1,0].grid(linestyle='--')


#子图4：中国123产业对比
China_first_industry=list(Chinese_industry['第一产业百分比'])
China_second_industry=list(Chinese_industry['第二产业百分比'])
China_third_industry=list(Chinese_industry['第三产业百分比'])
date2=list(Chinese_industry['指标名称'])

ax[1,1].plot(date2,China_first_industry,
            marker=markers[0], markerfacecolor=colors[0], markersize=4,markeredgewidth=0.5,
         color="k", linewidth=0.8, linestyle="-")
ax[1,1].plot(date2,China_second_industry,
            marker=markers[2], markerfacecolor=colors[2], markersize=4,markeredgewidth=0.5,
         color="k", linewidth=0.8, linestyle="-")
ax[1,1].plot(date2,China_third_industry,
            marker=markers[1], markerfacecolor=colors[1], markersize=4,markeredgewidth=0.5,
         color="k", linewidth=0.8, linestyle="-",label=labels[1])
ax[1,1].set_xlabel('年份',fontsize=24)
ax[1,1].set_ylabel('中国各产业百分比',fontsize=24)
plt.legend(['第一产业百分比','第二产业百分比','第三产业百分比'],
    loc='lower left',edgecolor='none',facecolor='none',fontsize=18)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
ax[1,1].grid(ls='--')
ax[1,1].spines['right'].set_color('none')
ax[1,1].spines['top'].set_color('none')

plt.tight_layout()
plt.savefig(r'C:\Users\PP\Desktop\数据\数据\国家禀赋对比.png')
plt.show()
```

```python

df1=list(df['超过100万城市群人口占比'])
df2=list(df['GDP/人'])
df21=[x*0.1 for x in df2]
```
